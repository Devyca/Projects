# To plot the Planck function of a single star as a function of temperature and frequency.

```python
import numpy as np
import matplotlib.pyplot as plt
from astropy.constants import c, h, k_B
from astropy import units as u
```

```python
# Constants
h = h.value  # Planck constant in J*s
c = c.value  # Speed of light in m/s
k = k_B.value  # Boltzmann constant in J/K
```

### Defining temperature and frequency parameters

```python
T = 3000* u.K
f = np.linspace(1e12, 15e15,1000)* u.Hz
```

###  Planck Function

The Planck function describes the spectral radiance of black-body radiation and is given by the formula:

$ B(\nu, T) = \frac{2h\nu^3}{c^2 \left(e^{\frac{h\nu}{k_B T}} - 1\right)} $

Where:
- \( B($\nu$, T) \) is the spectral radiance,
- \( $\nu$ \) is the frequency,
- \( T \) is the temperature,
- \( h \) is the Planck constant,
- \( c \) is the speed of light, and
- \( k_B \) is the Boltzmann constant.

To visualize the Planck function for different spectral types of stars, we plot this formula with varying temperatures and frequencies.


```python
def planck_fn(f, T):
    return (2*h*f**3)/(c**2*(np.exp((h*f)/(k*T))-1))
```

```python
B_f = planck_fn(f.value, T.value)  # Extract numerical values before calculations
```

```python
# Plotting
plt.figure(figsize=(10, 6))
# plt.loglog(f, B_f, label=f'T={T:.0f}')
plt.plot(f,B_f, label=f'T={T:.0f}')

# Adding labels and title
plt.xlabel('Frequency (Hz)')
plt.ylabel(r'$B_{\nu}(T)$, (eV/s/mÂ²/Sr/Hz)')
plt.title('Planck Function')
plt.legend()
plt.grid(True)
plt.show()
```

## To plot the Planck curves of different spectral types of stars

```python
# Spectral types and corresponding effective temperatures
spectral_types = ['O', 'B', 'A', 'F', 'G', 'K', 'M']
temp_range = {'O': 30000, 'B': 20000, 'A': 10000, 'F': 7000, 'G': 5500, 'K': 4000, 'M': 3000}

# Frequency range
f = np.linspace(1e12, 15e15, 1000)

# Plotting
plt.figure(figsize=(12, 8))

for spectral_type in spectral_types:
    T = temp_range[spectral_type]
    B_f = (2 * h * f**3) / (c**2 * (np.exp((h * f) / (k * T)) - 1))
    plt.plot(f, B_f, label=f'Spectral Type {spectral_type} (T={T:.0f})')

# Adding labels and title
plt.xlabel('Frequency (Hz)')
plt.ylabel(r'$B_{\nu}(T)$, (eV/s/mÂ²/Sr/Hz)')
plt.title('Planck Function for Different Spectral Types')
plt.legend()
plt.grid(True)
plt.show()
```

#### log - log plot

```python
# Frequency range
f = np.linspace(1e12, 15e16, 1000)

for spectral_type in spectral_types:
    T = temp_range[spectral_type]
    B_f = (2 * h * f**3) / (c**2 * (np.exp((h * f) / (k * T)) - 1))
    plt.loglog(f, B_f, label=f'Spectral Type {spectral_type} (T={T:.0f})')

# Adding labels and title
plt.xlabel('Frequency (Hz)')
plt.ylabel('Intensity')
plt.title('Planck Function for Different Spectral Types')
plt.legend()
plt.grid(True)
plt.show()
```

## Planck Function in Terms of Wavelength

The Planck function, expressed in terms of wavelength ($ \lambda$ \), is given by the formula:

$ B(\lambda, T) = \frac{2hc^2}{\lambda^5} \cdot \frac{1}{e^{\frac{hc}{\lambda k_B T}} - 1} $

Where:
- \( B($\lambda$, T) \) is the spectral radiance,
- \( $\lambda$ \) is the wavelength,
- \( T \) is the temperature,
- \( h \) is the Planck constant,
- \( c \) is the speed of light, and
- \( k_B \) is the Boltzmann constant.

To visualize the Planck function for different spectral types of stars, with varying temperatures and wavelengths.


```python
import numpy as np
import matplotlib.pyplot as plt
from astropy.constants import c, h, k_B

# Constants
h = h.value  # Planck constant in J*s
c = c.value  # Speed of light in m/s
k = k_B.value  # Boltzmann constant in J/K

# Spectral types and corresponding effective temperatures
spectral_types = ['O', 'B', 'A', 'F', 'G', 'K', 'M']
temp_range = {'O': 30000, 'B': 20000, 'A': 10000, 'F': 7000, 'G': 5500, 'K': 4000, 'M': 3000}

# Wavelength range (in meters)
wavelength = np.linspace(1e-8, 1e-6, 1000)

# Plotting
plt.figure(figsize=(12, 8))

for spectral_type in spectral_types:
    T = temp_range[spectral_type]
    frequency = c / wavelength
    B_wavelength = (2 * h * frequency**3) / (c**2 * (np.exp((h * frequency) / (k * T)) - 1))
    plt.plot(wavelength * 1e9, B_wavelength, label=f'Spectral Type {spectral_type} (T={T:.0f})')

# Adding labels and title
plt.xlabel('Wavelength (nm)')
plt.ylabel(r'$B_{\lambda}(T)$, (eV/s/mÂ²/Sr/Hz)')
plt.title('Planck Function for Different Spectral Types')
plt.legend()
plt.grid(True)
plt.show()
```
